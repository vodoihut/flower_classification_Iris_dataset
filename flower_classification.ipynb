```python
import pandas as pd
```

```python
## Đường dẫn tới tệp CSV
path = "/content/drive/MyDrive/Colab Notebooks/iris.csv"
## Đọc dữ liệu từ tệp CSV
data = pd.read_csv(path)
# Hiển thị 15 dòng đầu tiên của dữ liệu
data.head(15)
```

```python
# Xem thông tin về cột và kiểu dữ liệu
data.info()
```

```python
# Tóm tắt thống kê cơ bản về dữ liệu
data.describe()
```

```python
from sklearn.model_selection import train_test_split

# Tạo biến X chứa các đặc trưng (features) của dữ liệu
X = data[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]

# Tạo biến y chứa nhãn (labels) của dữ liệu
y = data['Name']

## Phân chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3)

# In ra 5 dòng đầu tiên của tập huấn luyện (đặc trưng)
print(X_train.head(5))
# In ra 5 dòng đầu tiên của tập huấn luyện (nhãn)
print(y_train.head(5))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Tạo một mô hình Random Forest với các tham số:
# - max_features: Số lượng đặc trưng tối đa được sử dụng cho việc chọn đặc trưng trong mỗi cây con.
# - n_estimators: Số lượng cây quyết định.
model = RandomForestClassifier(max_features=2, n_estimators= 150)

# Sử dụng tập huấn luyện huấn luyện mô hình
model.fit(X_train, y_train)
# Dự đoán kết quả trên tập kiểm tra của Mô hình Random Forest(Có thể không cần)
y_pred = model.predict(X_test)
#in ra các dự đoán được lưu trong biến y_pred
y_pred
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics

# Tạo một mô hình AdaBoostClassifier với các tham số:
# n_estimators: Số lượng cây quyết định trong AdaBoost.
# estimator: Mô hình cơ sở được sử dụng, trong trường hợp này, sử dụng mô hình đã tạo trước đó là Random Forest.
# learning_rate: Tốc độ học (learning rate) của AdaBoost.
# random_state: kiểm soát đối với tính ngẫu nhiên trong quá trình huấn luyện mô hình
 #(Nếu không đặt random_state thì scikit-learn sẽ sử dụng một giá trị mặc định random_state mà thư viện tự động tạo.)

ab_model = AdaBoostClassifier(n_estimators=50, estimator= model, learning_rate=1, random_state=42)
#Sử dụng tập huấn luyện huấn luyện mô hình
ab_model.fit(X_train, y_train)
```

```python
# Dự đoán kết quả sử dụng mô hình Random Forest kết hợp AdaBoost trên tập kiểm tra (X_test)
ab_predictions = ab_model.predict(X_test)
```

```python
# Đánh giá hiệu suất của mô hình Random Forest kết hợp AdaBoost
ab_accuracy = metrics.accuracy_score(y_test, ab_predictions)
print("Accuracy of AdaBoost with Random Forest:", ab_accuracy)
```
